- Description of TFIDIF formula ($\LaTeX$ OK): *Aid from Claude in converting Pythonic equations to mathematical ones and writing $\LaTeX$*

$\text{TF}(d,t) = \frac{1}{2} + \frac{1}{2} \cdot \frac{\mathbf{DTCM}^\top}{\max(\mathbf{DTCM}^\top)}$

$\text{DF}(t) = \sum_{d \in D} \mathbf{1}[\text{DTCM}_{d,t} > 0]$

$\text{IDF}(t) = \log_2\left(\frac{1 + N}{1 + \text{DF}(t)}\right) + 1$

$\text{TFIDF}(d,t) = \text{TF}(d,t) \cdot \text{IDF}(t)$

So: 

$\text{TFIDF}(d,t) = (\frac{1}{2} + \frac{1}{2} \cdot \frac{\mathbf{DTCM}^\top}{\max(\mathbf{DTCM}^\top)}) \cdot (\log_2\left(\frac{1 + N}{1 + \text{DF}(t)}\right) + 1)$
